**Цель:** установить параметры, влияющие на рыночную стоимость объектов недвижимости

Откровенных дублей, которые могли бы значительно повлиять на анализ, не выявлено. 
Для 49 объектов не указано название населённого пункта, заменим пропущенные значения на "название не указано"

```python
realty['locality_name'] = realty['locality_name'].fillna(value='название не указано')
print('Пропущенных значений в столбце locality_name:', realty['locality_name'].isnull().sum())
#Стандартизируем названия путём исключения буквы "ё"
realty['locality_name'] = realty['locality_name'].str.replace('ё', 'е')
```

Столбец is_apartment имеет 50 значений True, более 2000 значений False и более 20 000 значений не определенных. Так как апартаменты распространены менее, чем обычные квартиры, можно предположить, что если данные не были указаны, то недвижимость не является апартаментами. Заменим все пропущенные значения на False и переведем тип данных столбца в Булевый.

```python
realty['is_apartment'] = realty['is_apartment'].fillna(value=False)
realty['is_apartment'].astype(bool)
print('Является ли недвижимость апартаментами?')
display(realty['is_apartment'].value_counts())
```

Квартир без указания общего количества этажей в здании всего 0,4%, что не сильно влияет на общую картину. Эти данные лучше исключить из анализа, т.к. параметр важный для анализа, и если заменить значения на общее количество этажей в здании, то получится, что все эти квартиры будут автоматически попадать в категорию "на последнем этаже".

```python
realty = realty.dropna(subset=['floors_total']).reset_index(drop=True)
realty['floors_total'] = realty['floors_total'].astype(int)
```

Минимальной считается высота полотка 2,4 м, максимальная в домах обычно не превышает 5 м. Соответственно, все записи о высоте полотка выше 5 м или ниже 2,4 м можно считать ошибочными. Проверим, как распределяются высокие значения и сколько низких значений

```python
#Сделаем временную переменную, чтобы было удобно построить гистрограмму выбивающихся значений
sample_ceiling_high = realty.query('5 < ceiling_height')

#Изучим, сколько квартир имеют потолок выше 5 м или ниже 2.4 м
print('Объектов недвижимости с указанным значением высоты потолка больше 5 м:', len(sample_ceiling_high))
print('Объектов недвижимости с указанным значением высоты потолка меньше 2.4 м:', len(realty.query("ceiling_height < 2.4")))

sample_ceiling_high['ceiling_height'].hist(bins=100)
plt.title("Распределение значений высоты потолка")
plt.show()
```

Пик значений в диапазоне 20-30 м можно объяснить неправильной записью порядка значений и разделить значения на 10. Для высот от 5 до 20 м и для 100 м такое условие не подойдет, поэтому поскольку этих значений мало, их проще всего приравнять к медианному значению высоты потолка по всей выборке.
Все пропущенные значения в столбце заменим с учетом этажности зданий

```python
#Заменим значения в диапазоне от 20 до 50 м на те же, только на порядок меньше
realty.loc[(realty['ceiling_height'] > 20) & (realty['ceiling_height'] < 50), 'ceiling_height'] = realty.loc[(realty['ceiling_height'] > 20) & (realty['ceiling_height'] < 50), 'ceiling_height'] / 10
#Посчитаем медианное значение высоты потолка для выборки от 2.4 м до 5 м
ceiling_median = realty.query('2.4 <= ceiling_height <= 5')['ceiling_height'].median()
#Заменим значения слишком высоких или слишком низких потолков на медианное значение
realty.loc[(realty['ceiling_height'] > 5) | (realty['ceiling_height'] < 2.4), 'ceiling_height'] = ceiling_median
#Заполним все пропущенные значения медианными, в зависимости от этажности зданий
realty['ceiling_height'] = realty['ceiling_height'].fillna(realty.groupby('floors_total')['ceiling_height'].transform('median'))
#Для тех объектов, где нет аналогов по количеству этажей, заменим на медианное значение
realty['ceiling_height'] = realty['ceiling_height'].fillna(ceiling_median)
#Проверим полученные данные 
print('Высота потолка')
print(realty['ceiling_height'].describe())

realty['ceiling_height'].hist(bins=100)
plt.title("Распределение значений высоты потолка")
plt.show()
```

Теперь данные по высоте потолка выглядят вполне логично.
Для экономии пространства можно преобразовать этот столбец в целочисленный, с указанием высоты потолков в сантиметрах

```python
realty['ceiling_height'] = realty['ceiling_height'] * 100
realty = realty.rename(columns={'ceiling_height': 'ceiling_height_cm'})
realty['ceiling_height_cm'] = realty['ceiling_height_cm'].astype(int)
```

```python
display(realty.query('27 < living_area < 37')['last_price'].describe())
```

Такие квартиры стоят от 450 тыс., но в среднем 4,6 млн. рублей. Учитывая, что наша квартира на последнем этаже и имеет довольно низкое отношение жилой площади к общей, предположим, что составители базы данных ошиблись на 2 порядка. Заменим значение

Если не указано количество балконов, то можно предположить, что их в квартире не имеется. Заменим все пропущенные значения на 0 и переведем формат столбца в целочисленный

```python
realty['balcony'] = realty['balcony'].fillna(value=0).astype(int)
```

Обратим внимание на показатель жилой площади, где пропущено 1870 значений. Наиболее вероятно, этот показатель должен быть связан  с общей площадью помещения. Проверим эту гипотезу

```python
#Диаграмма рассеивания для параметров общей площади и жилой площади
realty.plot(x='total_area', y='living_area', kind='scatter')
plt.title("Соотношение жилой и общей площади помещения")
plt.xlabel("Общая площадь, кв. м")
plt.ylabel("Жилая площадь, кв. м")
plt.show()

print('Коэффициент коореляции для общей и жилой площади составляет', round(realty['total_area'].corr(realty['living_area']), 2))
living_total_rel = (realty['living_area'].median())/(realty['total_area'].median())
#Посчитаем, какую долю от общей занимает в подавляющем количестве случаев жилое пространство
print('В среднем жилая площадь занимает {}% от общей'.format(round(living_total_rel * 100)))
```

Заменим пропущенные значения жилой площади на долю от общей площади

```python
def fill_living(realty):
    living_area = realty['living_area']
    total_area = realty['total_area']    
    if pd.isna(living_area):
        living_area = total_area * 0.58
        return living_area
    return living_area
    
realty['living_area'] = realty.apply(fill_living, axis=1)
```

Теперь обработаем пропущенные данные о площади кухни. Несмотря на то, что в большинстве случаев общая площадь больше, чем сумма площадей жилой и кухни, расчитать насколько именно достаточно проблематично. Поэтому заменим пропущенные значения приблизительными данными, считая площадь кухни как разницу между общей площадью и жилой

```python
def fill_kitchen(realty):
    living_area = realty['living_area']
    total_area = realty['total_area']
    kitchen_area = realty['kitchen_area']
    if pd.isna(kitchen_area):
        kitchen_area = total_area - living_area
        return kitchen_area
    return kitchen_area
    
realty['kitchen_area'] = realty.apply(fill_kitchen, axis=1)
```

Пропущенные значения в столбцах "парки поблизости" и "пруды поблизости" можно заменить на 0, предположив, что если информация о них не указана, значит их поблизости нет. Также данные можно перевести в целочисленный формат, так как доли метров не влияют на результат при диапазоне значений 0-3000 м.

```python
realty['ponds_around3000'] = realty['ponds_around3000'].fillna(0).astype(int)
realty['parks_around3000'] = realty['parks_around3000'].fillna(0).astype(int)
```

Подозрительно выглядят кваритры, которые продаются очень быстро. Можно предположить, что так продаются наиболее дешевые квартиры. Категоризируем значения по дням и проверим, так ли это.

```python
#Функция для категоризации данных по скорости продажи квартир
def exposition_category_func(days_exposition):
    if (days_exposition <= 7):
        return '(00)<7'
    elif (days_exposition > 7 and days_exposition <= 30):
        return '(02)30-60'    
    elif (days_exposition > 30 and days_exposition <= 60):
        return '(02)30-60'
    elif (days_exposition > 60 and days_exposition <= 90):
        return '(03)90-90'
    elif (days_exposition > 90 and days_exposition <= 180):
        return '(04)90-180'
    elif (days_exposition > 180 and days_exposition <= 360):
        return '(05)180-360'
    else:
        return '(06)>360'
    
#Применение функции в новом столбце days_exposition_cat
exp_realty['days_exposition_cat'] = exp_realty['days_exposition'].apply(exposition_category_func)
#Вывод на экран средних значений, сгруппированных по days_exposition_cat
exp_realty_cat = exp_realty.groupby(['days_exposition_cat']).median().sort_values(by='days_exposition_cat')
display(exp_realty_cat)
```

### Стоимость жилья в наиболее популярных населенных пунктах

```python
#Группировка значения средней цены квартиры по названию населенного пункта
local_pivot = arealty.pivot_table(index='locality_name', values='meter_price', aggfunc=['count', 'median'])
local_pivot.columns = ['count', 'meter_price_median']
#Выбер 10 наиболее часто встречающихся населенных пунктов
local_pivot = local_pivot.sort_values(by='count', ascending=False).head(10).reset_index()
display(local_pivot)
#Расчет максимального и минимального значения средней цены за квадратный метр в населенных пунктах
local_max = local_pivot.loc[local_pivot['meter_price_median'] == local_pivot['meter_price_median'].max()].squeeze()
local_min = local_pivot.loc[local_pivot['meter_price_median'] == local_pivot['meter_price_median'].min()].squeeze()

print('Наибольшая средняя цена квардратного метра по выборке в {}e и составляет {} рублей'.format(local_max['locality_name'], round(local_max['meter_price_median'])))
print('Наименьшая средняя цена квардратного метра по выборке в {}e и составляет {} рублей'.format(local_min['locality_name'], round(local_min['meter_price_median'])))
```
